```python
import numpy as np
n = 5
x = np.arange(n)
print(x)
```

```python
y = np.random.rand(n)
print(y)
```

```python
import pandas as pd
import time
c = 10

df = pd.DataFrame({'y': y}, index=x)

start = time.time()
df["newy"] = df["y"] + c
end = time.time()
print(f"elapsed={end-start}")
print(df)
display(df)
```
